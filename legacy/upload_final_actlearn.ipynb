# Parameter-Information Active Learning for Non-Stationary GSMs

This notebook is a clean driver for the package in `src/actlearn_gsm`. The implementation lives in importable modules so the experiment can be tested, run from the CLI, and reviewed like a normal Python project.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
SRC = ROOT / "src"
if SRC.exists() and str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))
```

```python
from actlearn_gsm.config import (
    ActiveLearningConfig,
    BaselineConfig,
    EnsembleConfig,
    ExperimentConfig,
    GridConfig,
)
from actlearn_gsm.experiment import run_experiment
```

## Configuration

The default configuration matches the full experiment. Set `FAST_DEMO = True` to run a tiny CPU-only check from the notebook.

```python
FAST_DEMO = False

if FAST_DEMO:
    config = ExperimentConfig(
        grid=GridConfig(n_lon=8, n_lat=8, seed=42),
        ensemble=EnsembleConfig(
            size=2,
            num_inducing=4,
            num_mixtures=2,
            map_iters=2,
            lr=0.03,
        ),
        baseline=BaselineConfig(num_mixtures=2, iters=2, lr=0.05),
        active_learning=ActiveLearningConfig(
            num_init=10,
            num_steps=1,
            plot_every=1,
            seed=123,
        ),
        artifact_dir=Path("artifacts"),
        make_plots=False,
        device="cpu",
    )
else:
    config = ExperimentConfig(artifact_dir=Path("artifacts"), device="auto")

config
```

## Run

```python
results = run_experiment(config)
```
